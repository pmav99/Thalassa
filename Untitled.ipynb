# PyPoseidon stations format

```python
import meshzoo
import numpy as np
import pandas as pd
import xarray as xr

np.set_printoptions(linewidth=200)
```

### User Input

```python
grid_size = 4   # used by meshzoo
no_periods = 4
no_levels = 3
```

### Calculations

```python
# create grid
bary, cells = meshzoo.triangle(grid_size)
no_cells = len(cells)
cells
```

```python
# calculate lon/lat 
corners = np.array(
    [
        [0.0, -0.5 * np.sqrt(3.0), +0.5 * np.sqrt(3.0)],
        [1.0, -0.5, -0.5],
    ]
)
points = np.dot(corners, bary).T
lons = points[:, 0]
lats = points[:, 1]
```

```python
# get node indices
nodes=  np.unique(cells)
no_nodes = len(nodes)

# get timestamps
timestamps = pd.date_range("2001-01-01", freq="D", periods=no_periods)
```

```python
# Create elevation data
elevation_data = np.full((no_levels, no_periods, no_nodes), np.nan)
for level in levels:
    elevation_data[level, level:, :] = np.random.random((no_periods - level, no_nodes))
```

```python
observation_data = np.random.random((no_periods, no_nodes))
observation_data
```

```python
ds = xr.Dataset(
    coords=dict(
        node=nodes,
        time=timestamps,
        cell=range(no_cells),
        level=range(no_levels)
    ),
    data_vars=dict(
        lon=("node", lons), 
        lat=("node", lats), 
        elevation=(("level", "time", "node"), elevation_data),
        observation=(("time", "node"), observation_data),
        cell_nodes=(("cell", "node_per_face"), cells),
        # Additional node metadata
        station_name=("node", range(no_nodes)),
        station_id=("node", range(no_nodes)),
    ),
)
ds
```
